# Manual - MLflow

esse notebook mostra como o time de AI/RAG deve registrar experimentos no MLflow.
se o servidor já está rodando e configurado, basta só seguir os passos abaixo.

```python
import mlflow
import os

# configuração já vem do ambiente, não precisa alterar
os.environ['MLFLOW_S3_ENDPOINT_URL'] = 'http://minio:9000'
os.environ['AWS_ACCESS_KEY_ID']      = 'minioadmin'
os.environ['AWS_SECRET_ACCESS_KEY']  = 'minioadmin123'

mlflow.set_tracking_uri('http://mlflow:5000')
mlflow.set_experiment('triageai-training')

print('Conectado no MLflow:', mlflow.get_tracking_uri())
```

```python
# exemplo de como registrar um experimento
# substitui os valores pelos do seu modelo real

with mlflow.start_run(run_name='meu-experimento'):

    # parâmetros do modelo (configurações usadas no treino)
    mlflow.log_param('model_type', 'logistic_regression')
    mlflow.log_param('dataset', 'gold')
    mlflow.log_param('max_iter', 100)

    # métricas de avaliação
    mlflow.log_metric('accuracy',  0.85)
    mlflow.log_metric('f1_score',  0.82)
    mlflow.log_metric('precision', 0.80)
    mlflow.log_metric('recall',    0.84)

    print('experimento registrado!')
    print('Disponivel em: http://localhost:5000')
```

## onde ver os resultados

com o ambiente no ar (`make up`), acesse:

- **MLflow UI** → http://localhost:5000 (experimentos, métricas, modelos)
- **MinIO Console** → http://localhost:9001 (artefatos salvos, bucket `mlflow-artifacts`)

os modelos salvos com `mlflow.sklearn.log_model()` ficam automaticamente no bucket `mlflow-artifacts` do MinIO.
